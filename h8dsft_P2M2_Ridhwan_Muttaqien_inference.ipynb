```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model, load_model
from tensorflow.keras.layers import Input, Dense, Dropout

from google.colab import drive
drive.mount('/content/drive')
```

```python
model_path='/content/drive/MyDrive/coffee_predict.h5'
```

Karena ukuran model yang cukup besar maka model disimpan ke google drive. Berikut adalah link google drive nya: https://drive.google.com/file/d/1HfE_oeYk5ubKxQ6AWgRvduib_oltWE5E/view?usp=sharing

```python
model = load_model(model_path)
```

```python
def prediksi(file):
  img = tf.keras.utils.load_img(file, target_size=(224, 224))

  plt.imshow(img)
  x = tf.keras.utils.img_to_array(img)/255
  x = np.expand_dims(x, axis=0)

  images = np.vstack([x])
  classes = model.predict(images, batch_size=10)
  idx = np.argmax(classes)
  roast = ['dark','green','light','medium']
  print('Prediction is a {}'.format(roast[idx]))
```
